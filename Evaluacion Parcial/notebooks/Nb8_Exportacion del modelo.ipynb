```python
# Vamos a tomar los archivos que guardamos en Nb5 y Nb6 para armar el archivo modelo_coral_kmeans_k4.pkl
# Este archivo va a contener el KNN Imputer, el archivo con las alturas, y el modelo con K=4

import joblib
from pathlib import Path

carpeta_modelo = Path("modelo_exportado")

knn_imputer = joblib.load(
    carpeta_modelo / "knn_imputer_dataset_c.pkl"
)

artefacto_clustering = joblib.load(
    carpeta_modelo / "clustering_kmeans_k4.pkl"
)

modelo_final = {
    "kmeans": artefacto_clustering["kmeans"],
    "knn_imputer": knn_imputer,
    "alturas_modelo": artefacto_clustering["alturas_modelo"],
    "metadata": {
        **artefacto_clustering["metadata"],
        "imputacion": "KNNImputer",
        "salida": "cluster por perfil",
        "interpretacion": "clusters estadísticos sin interpretación física directa"
    }
}

joblib.dump(
    modelo_final,
    carpeta_modelo / "modelo_coral_kmeans_k4.pkl"
)

archivo_final = carpeta_modelo / "modelo_coral_kmeans_k4.pkl"

print("Modelo final guardado:", archivo_final.exists())
print("Ruta:", archivo_final.resolve())
print("Claves:", modelo_final.keys())
print("Cantidad de alturas:", len(modelo_final["alturas_modelo"]))
```
