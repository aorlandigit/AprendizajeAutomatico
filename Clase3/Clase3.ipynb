```python
# 2. Crear un gráfico de dispersión para visualizar las relaciones entre los pares de características numéricas `mpg` y `weight`.
import pandas as pd
import matplotlib.pyplot as plot
dset_automoviles_csv = pd.read_csv(r"C:\Users\orlan\OneDrive\Documents\Tecnicatura\2A1C - Aprendizaje Automatico\Semana 3\Dataset\automobile.csv", encoding= "ISO-8859-1")

plot.rcParams["figure.figsize"] = (8,5)
plot.style.use("ggplot")

print("\n \033[1m #2 Gráfico de Dispersion (scatter) Mpg vs Weight \033[0m \n")
plot.suptitle('Mpg vs Weight')
plot.title('Gráfico de Dispersion')
plot.xlabel('Mpg')
plot.ylabel('Weight')
plot.scatter(dset_automoviles_csv['mpg'],dset_automoviles_csv['weight'])
plot.show()
```

```python
# 3. Generar un histograma del consumo de combustible (`mpg`) para analizar su distribución.
import pandas as pd
import matplotlib.pyplot as plot
import math

dset_automoviles_csv = pd.read_csv(r"C:\Users\orlan\OneDrive\Documents\Tecnicatura\2A1C - Aprendizaje Automatico\Semana 3\Dataset\automobile.csv", encoding= "ISO-8859-1")

plot.rcParams["figure.figsize"] = (8,5)
plot.style.use("ggplot")

print("\n \033[1m #3 Histograma de Consumo Mpg \033[0m \n")
fig, ax = plot.subplots()
plot.suptitle('Mpg')
plot.title('Histograma')
plot.xlabel('Mpg')
cant_datos=len (dset_automoviles_csv)
cant_barras=round (1 + 3.322 * math.log10(cant_datos)) #regla de Sturges
ax.hist(dset_automoviles_csv['mpg'], bins=cant_barras, linewidth=0.5, edgecolor="white")
plot.show()
```

```python
# 4. Graficar un diagrama de caja (boxplot) de las características `cylinders` y `origin` para identificar posibles valores atípicos.
import pandas as pd
import matplotlib.pyplot as plot

dset_automoviles_csv = pd.read_csv(r"C:\Users\orlan\OneDrive\Documents\Tecnicatura\2A1C - Aprendizaje Automatico\Semana 3\Dataset\automobile.csv", encoding= "ISO-8859-1")

plot.rcParams["figure.figsize"] = (8,5)
plot.style.use("ggplot")

print("\n \033[1m #4 Boxplot Cylinders según Origin \033[0m \n")
grupo_origen= dset_automoviles_csv.groupby('origin')['cylinders'].apply(list)
plot.boxplot(grupo_origen, tick_labels=grupo_origen.index)

plot.suptitle('Cylinders por Origin')
plot.title('Boxplot')
plot.ylabel('Cylinders')
plot.xlabel('Origin')
plot.show()
```

```python
# 5. Realizar un gráfico de barras que muestre la distribución de automóviles según su año de modelo (`model_year`) y origen (`origin`).
import pandas as pd
import matplotlib.pyplot as plot

dset_automoviles_csv = pd.read_csv(r"C:\Users\orlan\OneDrive\Documents\Tecnicatura\2A1C - Aprendizaje Automatico\Semana 3\Dataset\automobile.csv", encoding= "ISO-8859-1")

plot.rcParams["figure.figsize"] = (8,5)
plot.style.use("ggplot")

print("\n \033[1m #5 Boxplot Cylinders según Origin \033[0m \n")
tabla_barras = pd.crosstab(dset_automoviles_csv['model_year'], dset_automoviles_csv['origin'])
tabla_barras.plot(kind='bar')
plot.suptitle('Distribución de modelos por año y origen')
plot.title('Gráfico de Barras')
plot.xlabel('Año y Origen de Modelos')
plot.ylabel('Cantidad de Modelos')

plot.show()
```
